```python
#抓取整頁HTML
import requests
from bs4 import BeautifulSoup
response = requests.get("https://travel.ettoday.net/category/%E6%A1%83%E5%9C%92/")
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())  #輸出排版後的HTML內容
```

<h1>一、以HTML標籤及屬性搜尋節點</h1>

```python
#只搜尋第一個符合條件的HTML節點，傳入要搜尋的標籤名稱
result = soup.find("h3")
print(result)
```

```python
#搜尋網頁中所有符合條件的HTML節點，傳入要搜尋的HTML標籤名稱。如果要更明確的搜尋，可以利用關鍵字參數(Keyword Argument)指定其屬性值。
#由於執行結果可能會搜出許多的HTML內容，所以最後也可以利用limit關鍵字參數(Keyword Argument)限制搜尋的節點數量

result = soup.find_all("h3", itemprop="headline", limit=3)
print(result)
```

```python
#如果要同時搜尋多個HTML標籤，可以將標籤名稱打包成串列(List)後，傳入find_all()方法(Method)中即可

result = soup.find_all(["h3","p"],limit = 10)
print(result)
```

```python
#當某一節點下只有單個子節點時，可以利用BeautifulSoup套件(Package)的select_one()方法(Method)，選取子節點

result = soup.find("h3", itemprop="headline")
print(result.select_one("a"))
```

```python
#點下有多個子節點時，則使用select()方法(Method)，選取子節點

result = soup.find("div",itemprop = "itemListElement")
print(result.select("a"))
```

<h1>二、以CSS屬性搜尋節點</h1>

```python
#要依據HTML的css屬性來進行節點的搜尋，需使用 class_ 關鍵字參數(Keyword Argument)來進行css屬性值的指定

titles = soup.find("p",class_="summary")
print(titles)
```

```python
#搜尋網頁中符合指定的HTML標籤及css屬性值的所有節點

titles = soup.find_all("p",class_="summary",limit =3)
print(titles)
```

```python
#而如果單純只想要透過css屬性值來進行HTML節點的搜尋，則可以使用BeautifulSoup套件(Package)的select()方法(Method)

titles = soup.select(".summary",limit = 3)
print(titles)
```

<h1>三、搜尋父節點</h1>

```python
#以上皆為向下的搜尋節點方式，如果想要從某一個節點向上搜尋，
#則可以使用BeautifulSoup套件(Package)的find_parent()或find_parents()方法(Method)

result = soup.find("a",itemprop = "url")
parents = result.find_parents("h3")
print(parents)

#搜尋<a>標籤且itemprop屬性值為url的節點，接著透過find_parents()方法(Method)，向上搜尋<h3>標籤的父節點。
```

<h1>四、搜尋前、後節點</h1>

```python
#在同一層級的節點，想要搜尋前一個節點，可以使用BeautifulSoup套件(Package)的find_previous_siblings()方法

result = soup.find("h3",itemprop = "headline")
previous_node = result.find_previous_siblings("a")
print(previous_node)
```

```python
#在同一層級的節點，想要搜尋後一個節點，則使用find_next_siblings()方法(Method)

result = soup.find("h3",itemprop = "headline")
next_node = result.find_next_siblings("p")
print(next_node)
```

<h1>五、取得屬性值</h1>

```python
#想要爬取「ETtoday的旅遊雲」桃園景點首頁的標題連結。
#首先，利用find_all()方法搜尋網頁中所有<h3>標籤且itemprop屬性值為headline的節點，
#接著，透過for迴圈讀取串列(List)中的節點，
#由於<h3>標籤底下只有一個<a>標籤，所以可以利用BeautifulSoup套件的select_one()方法進行選取

titles = soup.find_all("h3",itemprop = "headline")
for title in titles:
    print(title.select_one("a"))
```

```python
#最後，利用get()方法(Method)取得href屬性值中的網址

titles = soup.find_all("h3",itemprop = "headline")
for title in titles:
    print(title.select_one("a").get("href"))
```

<h1>六、取得連結文字</h1>

```python
#要取得<a>標籤的連結文字，可以利用BeautifulSoup套件(Package)的getText()方法(Method)

titles = soup.find_all("h3",itemprop = "headline")
for title in titles:
    print(title.select_one("a").getText())
```
